```python
import numpy as np
import pandas as pd
pd.set_option('max_columns', 1000)
pd.set_option('max_row', 400)
```

```python
Bladder_cancer_train = pd.read_csv('Bladder Urothelial Carcinoma (TCGA, PanCancer Atlas)/Bladder_cancer_train.csv', index_col = 'Entrez_Gene_Id')
Brain_cancer_train= pd.read_csv('Brain Lower Grade Glioma (TCGA, PanCancer Atlas)/Brain_cancer_train.csv', index_col = 'Entrez_Gene_Id')
Breast_cancer_train= pd.read_csv('Breast Invasive Carcinoma (TCGA, PanCancer Atlas)/Breast_cancer_train.csv', index_col = 'Entrez_Gene_Id')
Glioblastoma_train = pd.read_csv('Glioblastoma Multiforme (TCGA, PanCancer Atlas)/Glioblastoma_train.csv', index_col = 'Entrez_Gene_Id')
head_and_neck_cancer_train = pd.read_csv('Head and Neck Squamous Cell Carcinoma (TCGA, PanCancer Atlas)/head_and_neck_cancer_train.csv', index_col = 'Entrez_Gene_Id')
Kidney_cancer_train= pd.read_csv('Kidney Renal Papillary Cell Carcinoma (TCGA, PanCancer Atlas)/Kidney_cancer_train.csv', index_col = 'Entrez_Gene_Id')
Liver_cancer_train= pd.read_csv('Liver Hepatocellular Carcinoma (TCGA, PanCancer Atlas)/Liver_cancer_train.csv', index_col = 'Entrez_Gene_Id')
Lung_ad_train = pd.read_csv('Lung Adenocarcinoma (TCGA, PanCancer Atlas)/Lung_ad_train.csv', index_col = 'Entrez_Gene_Id')
Lung_sc_train = pd.read_csv('Lung Squamous Cell Carcinoma (TCGA, PanCancer Atlas)/Lung_sc_train.csv', index_col = 'Entrez_Gene_Id')
Stomach_ad_train = pd.read_csv('Stomach Adenocarcinoma (TCGA, PanCancer Atlas)/Stomach_ad_train.csv', index_col = 'Entrez_Gene_Id')
```

```python
df = pd.concat([Bladder_cancer_train, 
                Brain_cancer_train.iloc[:,1:], 
                Breast_cancer_train.iloc[:,1:], 
                Glioblastoma_train.iloc[:,1:], 
                head_and_neck_cancer_train.iloc[:,1:], 
                Kidney_cancer_train.iloc[:,1:], 
                Liver_cancer_train.iloc[:,1:], 
                Lung_ad_train.iloc[:,1:], 
                Lung_sc_train.iloc[:,1:], 
                Stomach_ad_train.iloc[:,1:], 
               ], axis = 1)
```
